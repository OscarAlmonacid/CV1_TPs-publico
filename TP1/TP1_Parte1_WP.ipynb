```python
#para mostrar las imagenes en linea con el codigo
%matplotlib inline 
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img_color1 = cv.imread('white_patch/wp_red.png')
image=np.asarray(img_color1)
B,G,R = cv.split(image)
#plt.imshow(image[:,:,::-1])
#plt.show()
```

```python
maxr=np.amax(R)
maxb=np.amax(B)
maxg=np.amax(G)
print ("Los valores maximos de RGB son: ",maxr,maxg,maxb)
sr=float(255)/maxr
sg=float(255)/maxg
sb=float(255)/maxb
print ("Los parametros de correcion son:",sr,sg,sb)
R=R*sr
G=G*sg
B=B*sb
R=R.astype(int)
G=G.astype(int)
B=B.astype(int)
R=np.asarray(R)
G=np.asarray(G)
B=np.asarray(B)
imageWP=cv.merge([B,G,R])
#plt.imshow(imageWP[:,:,::-1])
#plt.show()

fig = plt.figure()
# Imagen original
ax1=plt.subplot(221)
ax1.imshow(image[:,:,::-1])
ax1.set_title('Original')

# Imagen WP
ax2=plt.subplot(222)
ax2.imshow(imageWP[:,:,::-1])
ax2.set_title('WP')
```

Se ralizaron varias pruebas con las distintas imagenes. Se concluye que con este algoritmo no hay mejoras cuando uno de los canales predomina cercano al maximo, aqui el factor de corrección no influye.
