```python
#para mostrar las imagenes en linea con el codigo
%matplotlib inline 
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
#Se cargan y muestran las imagenes originales
fig, ax = plt.subplots(1,3, figsize=(16,16))
img_color1 = cv.imread("coord_cromaticas/CoordCrom_1.png")
img_color2 = cv.imread('coord_cromaticas/CoordCrom_2.png')
img_color3 = cv.imread('coord_cromaticas/CoordCrom_3.png')

img_rgb1 = cv.cvtColor(img_color1, cv.COLOR_BGR2RGB)
img_rgb2 = cv.cvtColor(img_color2, cv.COLOR_BGR2RGB)
img_rgb3 = cv.cvtColor(img_color3, cv.COLOR_BGR2RGB)

ax[0].imshow(img_rgb1)
ax[1].imshow(img_rgb2)
ax[2].imshow(img_rgb3)

plt.show()
```

```python
fig, ax = plt.subplots(1,3, figsize=(16,16))
#Se separa la imagen en los 3 canales RGB, luego se obtiene la suma y se la divide a cada canal
##IMAGEN 1
img1_r, img1_g, img1_b = cv.split(img_rgb1)
img1_r = img1_r.astype("uint64")
img1_g = img1_g.astype("uint64")
img1_b = img1_b.astype("uint64")
img_suma = img1_g + img1_r + img1_b
img_suma = img_suma.astype("uint64")
canales = [img1_r, img1_g, img1_b]
new_canales = [img1_r/img_suma,img1_g/img_suma,img1_b/img_suma]
new_img1 = cv.merge(new_canales)
new_img1 = cv.normalize(new_img1, new_img1, 0, 255, cv.NORM_MINMAX)
new_img1 = new_img1.astype("uint8")

##IMAGEN 2
img2_r, img2_g, img2_b = cv.split(img_rgb2)
img2_r = img2_r.astype("uint64")
img2_g = img2_g.astype("uint64")
img2_b = img2_b.astype("uint64")
img2_suma = img2_g + img2_r + img2_b
img2_suma = img2_suma.astype("uint64")
canales2 = [img2_r, img2_g, img2_b]
new_canales2 = [img2_r/img2_suma,img2_g/img2_suma,img2_b/img2_suma]
new_img2 = cv.merge(new_canales2)
new_img2 = cv.normalize(new_img2, new_img2, 0, 255, cv.NORM_MINMAX)
new_img2 = new_img2.astype("uint8")

##IMAGEN 3
img3_r, img3_g, img3_b = cv.split(img_rgb3)
img3_r = img3_r.astype("uint64")
img3_g = img3_g.astype("uint64")
img1_b = img3_b.astype("uint64")
img3_suma = img3_g + img3_r + img3_b
img3_suma = img3_suma.astype("uint64")
canales3 = [img3_r, img3_g, img3_b]
new_canales3 = [img3_r/img3_suma,img3_g/img3_suma,img3_b/img3_suma]
new_img3 = cv.merge(new_canales3)
new_img3 = cv.normalize(new_img3, new_img3, 0, 255, cv.NORM_MINMAX)
new_img3 = new_img3.astype("uint8")

ax[0].imshow(new_img1)
ax[1].imshow(new_img2)
ax[2].imshow(new_img3)
plt.show()
```
